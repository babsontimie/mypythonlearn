```python
# Concatenate the two series - series_A = 10,20,30,40,50  and  series_B = 'A','B','C','D','E'
series_A = 10,20,30,40,50
myseries_a = pd.Series(series_A)
print(myseries_a)

series_B = 'A','B','C','D','E' #concatenate both the series
myseries_b = pd.Series(series_B)
print(myseries_b)
series_C = pd.concat([myseries_a,myseries_b], ignore_index=True)
print(series_C)
```
