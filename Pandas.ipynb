```python
# Visualization
# pip install matplotlib
# pip install seaborn

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# seaborn is the advanced version of matplotlib
```

```python

days = ['Moinday', 'Tuesday', 'Wednesday', 'Thursday','Friday','Saturday']
sales = [450,200,616,438,500,250]
plt.plot(days,sales, color='green',marker='*')  # o, D, v
plt.title('Line plot of Days and Sales')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.grid()
plt.show()
```

```python
# Bar plots
#Positive gradient
```

```python
plt.bar(days,sales,color='blue')
plt.title('Bar plot of Days and Sales')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.grid()
plt.show()
```

```python
#Giving colors to each bar
plt.bar(days,sales,color = ['red','green','blue','yellow','orange','purple'])
plt.title('Bar plot of Days and Sales')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.grid()
plt.show()
```

```python
# Changing to horizontal plot. Use barh
plt.barh(days,sales,color = ['red','green','blue','yellow','orange','purple'])
plt.title('Horizontal Bar plot of Days and Sales')
plt.xlabel('Sales Amount')
plt.ylabel('Days')
plt.grid()
plt.show()
```

```python
# Changing size of grid
plt.barh(days,sales,color = ['red','green','blue','yellow','orange','purple'])
plt.title('Horizontal Bar plot of Days and Sales')
plt.xlabel('Sales Amount')
plt.ylabel('Days')
plt.grid(True)
plt.show()
```

```python
# SCATTER PLOTS
roll_no = [101,102,103,104,105,106,107,108,109,110]
marks = [10,18,34,90,43,67,21,67,45,50]
plt.scatter(roll_no,marks, color = 'red',marker='*',s=200) # s is used to chnage the size of marker
plt.title('Scatter plot of Roll No and Marks', color='red',fontsize=20)
plt.xlabel('Roll Number of students')
plt.ylabel('Marks of the. students')
plt.grid(True)
plt.show()
```

```python
# create multiple lines in a single line plot

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday','Friday','Saturday']
sale1 = [50,60,40,90,35,84]
sale2 = [78,39,23,80,50,47]
sale3 = [56, 75,87,67,35,90]
plt.plot(days,sale1, color='green',marker='*',label='Sale1')
plt.plot(days,sale2,color='purple',marker='D'label='Sale2')
plt.plot(days,sale3,color='red',marker='v'label='Sale2')
plt.title('Line plot of Days and Sales')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()
```

```python
# Adding labeel to the line plot
plt.plot(days,sale1, color='green',marker='*',label='Sale1')
plt.plot(days,sale2,color='purple',marker='D',label='Sale2')
plt.plot(days,sale3,color='red',marker='v',label='Sale2')
plt.title('Line plot of Days and Sales')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.legend()  # The legend is used to ouput the labels
plt.show()
```
